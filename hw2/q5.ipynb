```python
import numpy as np
import pandas as pd
```

```python
suspects = ('gadzi', 'annie', 'ren')
utensils = ('spoon', 'fork', 'chopsticks')

# making everything uniform...
prior_probs = np.array([1/3, 1/3, 1/3])
gadzi_probs = np.array([1/3, 1/3, 1/3])
annie_probs = np.array([1/3, 1/3, 1/3])
ren_probs = np.array([1/3, 1/3, 1/3])
```

```python
def draw_prior(suspects, probs):
    return np.random.choice(suspects, p=probs)

def draw_model(utensils, suspect, gadzi_probs, annie_probs, ren_probs):
    if (suspect == 'gadzi'):
        return np.random.choice(utensils, p=gadzi_probs)

    elif (suspect == 'annie'):
        return np.random.choice(utensils, p=annie_probs)

    else:
        return np.random.choice(utensils, p=ren_probs)

def draw_joint(utensils, suspects, prior_probs, gadzi_probs, annie_probs, ren_probs):
    s = draw_prior(suspects, prior_probs)
    w = draw_model(utensils, s, gadzi_probs, annie_probs, ren_probs)
    return f"{s}_{w}"

def simulator(*args, num_sims=10000):
    return [draw_joint(*args) for _ in range(num_sims)]
```

```python
for i in range(5):
  num_sims = 10**i

  sims = simulator(utensils, suspects, prior_probs, gadzi_probs, annie_probs, ren_probs, num_sims=num_sims)

  scenarios, counts = np.unique(sims, return_counts=True)

  print(f"num_sims = {num_sims}:")
  for s, c in zip(scenarios, counts):
    print(f'{s} occurred with a probability of {round(c / num_sims, 2)}.')

  print("\n\n")
```

```python
# converting scenario counts into probabilities...
freqs = counts / num_sims
scenario_to_p = dict(zip(scenarios, freqs))

# making the table with rows being suspects and cols being utensils...
sim_joint = np.zeros((len(suspects), len(utensils)))

for i, s in enumerate(suspects):
  for j, w in enumerate(utensils):
    sim_joint[i, j] = scenario_to_p.get(f"{s}_{w}", 0.0)


# simulated joint table...
sim_df = pd.DataFrame(sim_joint, index=suspects, columns=utensils)
print(f"Simulated joint table (N={num_sims}):")
display(sim_df)
```

```python
# forming analytic joint table...
analytic_joint = np.full((len(suspects), len(utensils)), 1/9)
analytic_df = pd.DataFrame(analytic_joint, index=suspects, columns=utensils)

print("Analytic joint table:")
display(analytic_df)
```

```python
# calculating "error" btwn both tables...
abs_err_df = (sim_df - analytic_df).abs()

print("Absolute error |sim - analytic|:")
display(abs_err_df)

print("Max absolute error:", abs_err_df.to_numpy().max())
```
